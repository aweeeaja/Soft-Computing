```python
#Fungsi Keanggotaan Humidity
def humidity_rendah(x):
  if x <= 40:
    return 1
  elif 40 <= x <= 90:
    return (90 - x)/50
  else:
    return 0

def humidity_tinggi(x):
  return 2*((100-x)/(100-90))**2
```

```python
#Fungsi Keanggotaan Temperatur
def temp_dingin(x):
  return (24-x)/(24-0)
def temp_sedang(x):
  return (x-19)/(24-19)
def temp_panas(x):
  if x >= 24:
    return 1
  elif 33<= x <= 40:
    return (x-33)/(40-33)
  else:
    return 0
```

```python
#Fungsi Keanggotaan Intensitas
def intensitas_kurangi(x):
  z = 5 - ((5-0)*x)
  return z
def intensitas_tingkatkan(x):
  z = (x * (10-5)) + 5
  return z
```

```python
#Humidity dan temperature
humid = 50
temp = 26
```

```python
#Rule 1:  if Humidity Rendah and Temperatur dingin then Intensitas Kurangi
a1 = min(humidity_rendah(humid),temp_dingin(temp))
z1 = 10* temp + 100 #intensitas kurangi

#Rule 2:  if Humidity Rendah and Temperatur sedang then Intensitas tingkatkan
a2 = min(humidity_rendah(humid),temp_sedang(temp))
z2 = 10* temp + 150 #intensitas tingkatkan

#Rule 3:  if Humidity Rendah and Temperatur panas then Intensitas tingkatkan
a3 = min(humidity_rendah(humid),temp_panas(temp))
z3 = 10* temp + 200 #tingkatkan

#Rule 4:  if Humidity Tinggi and Temperatur dingin then Intensitas Kurangi
a4 = min(humidity_tinggi(humid),temp_dingin(temp))
z4 = 5*humid + 2* temp #kurangi

#Rule 5:  if Humidity tinggi and Temperatur sedang then Intensitas Kurangi
a5 = min(humidity_tinggi(humid),temp_sedang(temp))
z5 = 5*humid + 4* temp + 150#kurangi

#Rule 6:  if Humidity Tinggi and Temperatur panas then Intensitas tingkatkan
a6 = min(humidity_tinggi(humid),temp_panas(temp))
z6 = 5*humid + 5* temp + 450#tingkatkan
```

```python
defuz = (a1*z1 + a2*z2 + a3 * z3 + a4*z4 + a5*z5 + a6*z6)/(a1+a2+a3+a4+a5+a6) #Perhitungan Metode Average
print(f'Intensitas seharusnya {round(defuz):,}')
```
